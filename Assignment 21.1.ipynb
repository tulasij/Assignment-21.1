```python
from sklearn.datasets import load_boston
```

```python
from sklearn import linear_model
```

```python
data =load_boston()
```

```python
# define the data/predictors as the pre-set feature names  
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
# Put the target (housing value -- MEDV) in another DataFrame
target = pd.DataFrame(data.target, columns=["MEDV"])
```

```python
#we have the data frame that contains the independent variables (marked as “df”) and the data frame with the dependent variable 
#(marked as “target”). Let’s fit a regression model using SKLearn. First we’ll define our X and y — we will use all the variables in the data frame to predict the housing price:
        X = df
        y = target["MEDV"]
```

```python
#The lm.fit() function fits a linear model. 
lm = linear_model.LinearRegression()
model = lm.fit(X,y)
```

```python
#We want to use the model to make predictions so we’ll use lm.predict()
predictions = lm.predict(X)
```

```python
#The print function would print the first 5 predictions for y. Removing [0:5] would print the entire list):
print(predictions)[0:5]
```

```python
lm.score(X,y)#This is the R² score of our model.
```
